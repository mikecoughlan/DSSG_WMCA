```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/WMCA/data_cleaning.csv")
```

# Standardising entries

We will correct recurrent errors in the descriptions and translate/remove the Welsh in the entries.

```python
cat_var = df.select_dtypes(include= ['object','category']).columns.tolist()

# Not these variables
cat_var.remove('lmk-key')
cat_var.remove('address')
cat_var.remove('postcode')
```

```python
# remove bilingual 
for c in cat_var:
  df[c] = df[c].str.replace(r"\|(.*)", "")
```

```python
for c in ['floor-description','walls-description','roof-description']:
    # standardising the unit used
    df[c] = df[c].str.replace(r"W(.*?)K",'W/m²K')
    # removing nonsense 
    df[c] = df[c].str.replace("\*\*\* INVALID INPUT Code \: 57 \*\*\*","")
    # cleaning up 
    df[c] = df[c].str.replace(r'\.$','')
    df[c] = df[c].str.replace(r'\,$','')
    df[c] = df[c].str.replace("  \+"," +")
    df[c] = df[c].str.replace("[ \t]+$","")
    df[c] = df[c].str.replace('  ',' ')
    df[c] = df[c].str.replace('Average thermal transmittance 1 ','Average thermal transmittance 1.00 ')
    df[c] = df[c].str.replace('Average thermal transmittance =','Average thermal transmittance')
    df[c] = df[c].str.rstrip(' ')
```

```python
# removing phrases like as built which aren't adding anything
df['walls-description'] = df['walls-description'].str.replace("\(assumed\)","")
df['walls-description'] = df['walls-description'].str.replace("as built, ","")
df['walls-description'] = df['walls-description'].str.replace("cavity\.","cavity wall,")
# standardising punctuation
df['walls-description'] = df['walls-description'].str.replace("solid brick\.","solid brick,")
# standardising language
df['walls-description'] = df['walls-description'].replace(
    regex=["granite or whin,", "stone \(granite or whin\)\."],
    value="granite or whinstone,")
df['walls-description'] = df['walls-description'].replace(
    regex=["with external insulation", "with internal insulation", 
     "with additional insulation", "with insulation"],
     value="insulated")
# clean up
df['walls-description'] = df['walls-description'].str.replace('\+ chr\(13\) \+','+')
df['walls-description'] = df['walls-description'].str.replace("[ \t]+$","")
df['walls-description'] = df['walls-description'].str.replace('timber frame\.','timber frame,')
df['walls-description'] = df['walls-description'].str.replace('\?','')
df['walls-description'] = df['walls-description'].str.replace('system built\.','system built,')
```

```python
# standardising units
df['roof-description'] = df['roof-description'].str.replace(" mm",'mm')
df['roof-description'] = df['roof-description'].str.replace(">= 300mm",">=300mm")
df['roof-description'] = df['roof-description'].str.replace(">=300mm","300+mm")
# standardising descriptions
df['roof-description'] = df['roof-description'].str.replace("Roof room,","Roof room(s),")
df['roof-description'] = df['roof-description'].str.replace("another dwelling above","other premises above")
df['roof-description'] = df['roof-description'].replace(
    regex=["Other premises above","Dwelling Above"],
    value="(other premises above)")
df['roof-description'] = df['roof-description'].str.replace(r"0 W/m²K",r" W/m²K")
df['roof-description'] = df['roof-description'].str.replace("  W/m²K"," 0.0 W/m²K")
df['roof-description'] = df['roof-description'].str.replace("Roof room\(s\), no insulation\(assumed\)","Roof room(s), no insulation (assumed)")
df['roof-description'] = df['roof-description'].str.replace("\(assumed\)","")
# cleanup
df['roof-description'] = df['roof-description'].str.replace("mmmm","mm")
df['roof-description'] = df['roof-description'].str.replace(",mm",",")
```

```python
# removing assumed
df['floor-description'] = df['floor-description'].str.replace("\(assumed\)","")
# standardising language
df['floor-description'] = df['floor-description'].replace(
    regex=["\(another dwelling below\)", "\(other premises below\)"],
    value='other premises below') 
df['floor-description'] = df['floor-description'].str.replace("solid.",'solid,') 
df['floor-description'] = df['floor-description'].str.replace("uninsulated",'no insulation,') 
df['floor-description'] = df['floor-description'].str.replace("insulation=100mm",'100 mm insulation') 
df['floor-description'] = df['floor-description'].str.replace(", \(assumed\)",' (assumed)')
df['floor-description'] = df['floor-description'].str.replace("insulation=25mm",'25 mm insulation')
df['floor-description'] = df['floor-description'].str.replace("insulation=75mm",'75 mm insulation')
df['floor-description'] = df['floor-description'].str.replace("limited insulated",'limited insulation')
# cleanup
df['floor-description'] = df['floor-description'].str.replace(r'^, ','')
df['floor-description'] = df['floor-description'].str.replace(r'\?','')
```

```python
# Extract thermal transmittance feature
for c in ['floor-description','walls-description','roof-description']:
  col_name = c.split('-')[0] + "-thermal-transmittance"
  transmittance_col = df[c].str.findall(r'(\d.\d*) W/m²K')
  df[col_name] = round(transmittance_col.str[0].astype(float),2)
```

```python
construction_year = df['construction-age-band'].str.findall(r"(\d{4})")
df['construction-age-band'] = construction_year.str[-1].astype('datetime64')
```

```python
df['transaction-type'] = df['transaction-type'].str.replace(" - this is for backwards compatibility only and should not be used","")
df['transaction-type'] = df['transaction-type'].str.replace("Stock Condition Survey", "Stock condition survey")
df['transaction-type'] = df['transaction-type'].replace("not recorded", np.nan)
```

```python
df['main-fuel'] = df['main-fuel'].str.replace(" - this is for backwards compatibility only and should not be used","")
```

```python
df['tenure'] = df['tenure'].str.replace('Rented', 'rental')
df['tenure'] = df['tenure'].replace('Not defined - use in the case of a new dwelling for which the intended tenure in not known. It is no', np.nan)
df['tenure'] = df['tenure'].str.lower()
```

```python
df['floor-level'] = df['floor-level'].str.lstrip('0')
df['floor-level'] = df['floor-level'].replace(
    regex=['ground floor', ''], 
    value='Ground')
df['floor-level'] = df['floor-level'].replace(
    regex=[r'\dth', r'\dnd', r'\drd'], 
    value='')
df['floor-level'] = df['floor-level'].replace('Basement', '-1')
```

```python
df['glazed-area'] = df['glazed-area'].replace('Not Defined', np.nan)
```

```python
# standardising language
df['mainheatcont-description'] = df['mainheatcont-description'].str.lower()
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("programmer, no thermostat","programmer, no room thermostat")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("flat rate charging\*","flat rate charging")
df['mainheatcont-description'] = df['mainheatcont-description'].replace(
    regex=[r"\+",r"\&"] ,
    value="and")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("trv.s","trvs")
df['mainheatcont-description'] = df['mainheatcont-description'].replace(
    regex=["thermostats", " stat", "thermostatic"],
    value="thermostat")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("communit ","community ")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("to the use of community heating","to use of community heating")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("controls","control")
df['mainheatcont-description'] = df['mainheatcont-description'].replace(
    regex=["prog ","program "],
    value="programmer ")
df['mainheatcont-description'] = df['mainheatcont-description'].replace(
    regex=["programmerand",r"programmer\?and"],
    value="programmer and")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("roomthermostat","room thermostat")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("delayed start thermostat and program and trvs","delayed start thermostat, program and trvs")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("flat rate charging, programmer no room thermostat","flat rate charging, programmer, no room thermostat")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace(" 2 "," two ")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("roomstat","room thermostat")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("programmer and room thermostat and trvs","programmer, room thermostat and trvs")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("programmer and trvs and boiler energy manager","programmer, trvs and boiler energy manager")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("programmer and trvs and bypass","programmer, trvs and bypass")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("programmer and trvs and flow switch","programmer, trvs and flow switch")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("temp+$","temperature")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("no thermostat control of room temperature","no thermostat control")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("appliance thermostat and programmer","programmer and appliance thermostat")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("delayed start thermostat and programmer and trvs","delayed start thermostat, programmer and trvs")
df['mainheatcont-description'] = df['mainheatcont-description'].str.replace("no time or thermostat control of temperature","no time or thermostat control of room temperature")
```

```python
df['hotwater-description'] = df['hotwater-description'].str.replace(r"\|.*", '')
df['hotwater-description'] = df['hotwater-description'].str.replace("cylinder thermostat","cylinderstat")
df['hotwater-description'] = df['hotwater-description'].replace(
    regex=["No system present :", "No hot water system present -"],
    value="No system present:")
df['hotwater-description'] = df['hotwater-description'].replace(
    regex=["From community scheme", "community scheme"],
    value="Community scheme")
df['hotwater-description'] = df['hotwater-description'].str.replace("Community heat pump","Community scheme with CHP")
df['hotwater-description'] = df['hotwater-description'].replace(
    regex=["From secondary heater","From second main heating system"],
    value="From secondary system")
df['hotwater-description'] = df['hotwater-description'].str.replace("SAP05:Hot-Water","SAP:Hot-Water")
df['hotwater-description'] = df['hotwater-description'].str.replace("plus solar, no cylinderstat","no cylinderstat, plus solar")
df['hotwater-description'] = df['hotwater-description'].str.replace("none","No system present: electric immersion assumed")
df['hotwater-description'] = df['hotwater-description'].str.replace("plus solar, flue gas heat recovery","flue gas heat recovery, plus solar")
df['hotwater-description'] = df['hotwater-description'].str.replace("no cylinderstat, no cylinderstat","no cylinderstat")
df['hotwater-description'] = df['hotwater-description'].replace("***SAMPLE***",np.nan)
```

```python
df['windows-description'] = df['windows-description'].str.lower()
df['windows-description'] = df['windows-description'].str.replace("glazed","glazing")
df['windows-description'] = df['windows-description'].str.replace("fully","full")
df['windows-description'] = df['windows-description'].str.replace("single glazingsingle glazing","single glazing")
df['windows-description'] = df['windows-description'].str.replace("single glazingdouble glazing","single glazing and double glazing")
df['windows-description'] = df['windows-description'].str.replace("single glazingsecondary glazing","single glazing and secondary glazing")
df['windows-description'] = df['windows-description'].str.replace("  "," ")
```

```python
df['secondheat-description1'] = df.apply(lambda row: row['lighting-description'] if 'lighting' in str(row['secondheat-description']) else row['secondheat-description'],axis = 1)
df['lighting-description1'] = df.apply(lambda row: row['secondheat-description'] if 'lighting' in str(row['secondheat-description']) else row['lighting-description'],axis = 1)
df.drop(columns= ['secondheat-description','lighting-description'],axis=1,inplace=True)
df.rename(columns={'secondheat-description1':'secondheat-description','lighting-description1':'lighting-description'},inplace=True)
```

```python
## CONVERT TO NUMERICAL ##
df['lighting-description'] = df['lighting-description'].str.lower()
# Converting lighting descriptions in % of low energy lighting
df['lighting-description'] = df['lighting-description'].str.replace(r'low energy lighting in 120% of fixed outlets','low energy lighting in 100% fixed outlets')
df['lighting-description'] = df['lighting-description'].str.replace(r'low energy lighting 100% of fixed outlets','low energy lighting in 100% fixed outlets')
df['lighting-description'] = df['lighting-description'].str.replace(r'no low energy lighting','low energy lighting in 0% fixed outlets')
df['lighting-description'] = df['lighting-description'].str.replace(r'sap05:lighting','')
```

```python
# Extract % low energy lighting
df['percentage-low-energy-lighting'] = df['lighting-description'].str.extract('([-+]?(?:\d*\.\d+|\d+))')
df['percentage-low-energy-lighting'] = df['percentage-low-energy-lighting'].astype(float)
```

```python
df['secondheat-description'] = df['secondheat-description'].str.rstrip(', ')
df['secondheat-description'] = df['secondheat-description'].str.replace('LNG', 'LPG')
df['secondheat-description'] = df['secondheat-description'].str.replace('lpg', 'LPG')
df['secondheat-description'].replace(regex=[r'\,\s\(null\)', r'\s*\(assumed\)'], value='', inplace=True)
```

```python
df['main-heating-controls'] = df['main-heating-controls'].replace('%%MAINHEATCONTROL%%', np.nan)
df['main-heating-controls'] = df['main-heating-controls'].astype(float)
```
